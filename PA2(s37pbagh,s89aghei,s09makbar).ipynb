##Importing the required libraries

```python
from time import sleep
import os
```

##definition of a class implementing a Two dimension cellular automata

###Function __init__ :
this function takes the width and the height and the rule for implementing a 2 dimensional board. This function Also creates a log file to write the alive cells count in each iteration.

###Function step:
 this function implements each step of the cellular automata by applying the rule one each cell. The boundary is periodic which implementing by calculating the mod of the cordination of the neighbors.

###Function run:
this function takes number of iterations as an arguments and starts to do the steps. After Each step it prints the cellular automata, calculates the total number of alive cells and writes it to a log file. For a better rrepresenting after each step it weites for 100 miliseconds then clears the terminal windows.

###function __str__:
Gives a decent representing of the board in str format (# if cell is alive and a blank space if it is dead)

###function getNeighbors:
takes the position of a cell and returns its' neighbors in a list.

###functions glider, gospersGliderGun, blinker, Acron:
these functions initialize a pattern in the board with mentioned patterns in the exercise sheet. We decided to implement Acron pattern as our elective class 3 pattern.



```python

class TwoDCellularAutomata:
    def __init__(self, rule, width, height):
        self.rule = rule
        self.width = width
        self.height = height
        self.cells = [[0 for _ in range(width)] for _ in range(height)]

        with open('log.txt', 'w') as f:
            f.write('Start')
            f.write('\n')



    def step(self):
        new_cells = [[0 for _ in range(self.width)] for _ in range(self.height)]
        for i in range(1, self.height):
            for j in range(self.width):
                new_cells[i][j] = self.rule(self.cells[i][j], self.getNeighbors(i, j))
        self.cells = new_cells

    def run(self, steps):
        for i in range(steps):
            logger_file(self,i)
            print(self)
            sleep(0.1)
            self.step()
            os.system('cls' if os.name == 'nt' else 'clear')
    def totalAlive(self):
        return sum([sum(row) for row in self.cells])

    def getNeighbors(self, i, j):
        neighbors = []
        ys=i-1
        ye=(i+1)%self.height
        xs=j-1
        xe=(j+1)%self.width
        for y in range(ys, ye+1):
            for x in range(xs, xe+1):
                if x == j and y == i:
                    continue
                neighbors.append(self.cells[y][x])
        return neighbors
    def __str__(self):
        return '\n'.join([''.join(['#' if cell else ' ' for cell in row]) for row in self.cells])
    def glider(self):
        #glider pattern
        self.cells[0][1] = 1
        self.cells[1][2] = 1
        self.cells[2][0] = 1
        self.cells[2][1] = 1
        self.cells[2][2] = 1
    def gospersGliderGun(self):
        #Gosper's Glider Gun pattern
        self.cells[1][5] = 1
        self.cells[1][6] = 1
        self.cells[2][5] = 1
        self.cells[2][6] = 1
        self.cells[11][5] = 1
        self.cells[11][6] = 1
        self.cells[11][7] = 1
        self.cells[12][4] = 1
        self.cells[12][8] = 1
        self.cells[13][3] = 1
        self.cells[13][9] = 1
        self.cells[14][3] = 1
        self.cells[14][9] = 1
        self.cells[15][6] = 1
        self.cells[16][4] = 1
        self.cells[16][8] = 1
        self.cells[17][5] = 1
        self.cells[17][6] = 1
        self.cells[17][7] = 1
        self.cells[18][6] = 1
        self.cells[21][3] = 1
        self.cells[21][4] = 1
        self.cells[21][5] = 1
        self.cells[22][3] = 1
        self.cells[22][4] = 1
        self.cells[22][5] = 1
        self.cells[23][2] = 1
        self.cells[23][6] = 1
        self.cells[25][1] = 1
        self.cells[25][2] = 1
        self.cells[25][6] = 1
        self.cells[25][7] = 1
        self.cells[35][3] = 1
        self.cells[35][4] = 1
        self.cells[36][3] = 1
        self.cells[36][4] = 1
    def blinker(self):
        #A blinker pattern
        self.cells[(self.height//2)][self.width//2] = 1
        self.cells[(self.height//2)+1][self.width//2] = 1
        self.cells[(self.height//2)+2][self.width//2] = 1
    def acron(self):
        #Acron pattern which is an example of a class 3 pattern
        self.cells[0+(self.height//2)][1+(self.width//2)] = 1  # The single live cell in the first part
        self.cells[1+(self.height//2)][3+(self.width//2)] = 1  # The first of the pair
        self.cells[2+(self.height//2)][0+(self.width//2)] = 1  # The second of the pair
        self.cells[2+(self.height//2)][1+(self.width//2)] = 1  # First cell of the three in a row
        self.cells[2+(self.height//2)][4+(self.width//2)] = 1  # Second cell of the three in a row
        self.cells[2+(self.height//2)][5+(self.width//2)] = 1  # Third cell of the three in a row
        self.cells[2+(self.height//2)][6+(self.width//2)] = 1  # The isolated cell in the last column




```

##Definition of Conway's Game of Life rule
Here the rule is defined and in further steps this function will be passed to the game class.

```python
def ConwayRule(cell, neighbors):
    if cell == 1:
        if sum(neighbors) < 2 or sum(neighbors) > 3:
            return 0
        else:
            return 1
    else:
        if sum(neighbors) == 3:
            return 1
        else:
            return 0
```

##Logger function definition
This function adds each step of the run in a log file.

```python
def logger_file(game,step):
    with open('log.txt', 'a') as f:
        f.write(f'Step: {step} ')
        f.write(f'Total Alive: {game.totalAlive()}')
        f.write('\n')

```

##The main procedure of running the program
Asks the initial pattern from the user and runs the program.

```python
def main():
    game = TwoDCellularAutomata(ConwayRule, 101, 82)
    option = int(input("1. Glider\n2. Gosper's Glider Gun\n3. Blinker\n4. Acorn\npleasae select the option:"))
    if option == 1:
        game.glider()
    elif option == 2:
        game.gospersGliderGun()
    elif option == 3:
        game.blinker()
    elif option == 4:
        game.acron()

    game.run(10)

main()
```
